```python
import matplotlib.pyplot as plt
```

```python
# If no x-axis is given it takes the x values as indices

plt.plot([2,4,6,4])
plt.ylabel("Numbers")
plt.xlabel("Indices")
plt.title("Line Plot")
plt.show()
```

```python
import numpy as np
# plt.plot([1,2,3,4],np.array([1,4,9,16])) # it takes lists, numpy arrays all
# first is x axis , second list is y -axis
x = np.arange(0,100,2)
y = x**2
plt.plot(x,y,'ro') # r= red , o = dots instead of line
plt.ylabel("Squares")
plt.xlabel("Numbers")
plt.title("Line Plot")
plt.grid() # shows in grid easy to visualize
plt.show()
```

Multiple Lines and Legend

```python
import numpy as np
a = np.arange(0 , 5, 0.2)
print(a)

plt.plot(a, a**2, 'b--', label = '^2')
plt.plot(a, a**3, 'rs', label = '^3')
plt.plot(a, a**4, 'g^', label = '^4')
plt.legend()
plt.xlabel('Numbers')
plt.ylabel("Exponents")
plt.title("Multiple Lines")
plt.grid()
plt.show()
```

```python
# Linewidth property
plt.plot([1,2,3,4],np.array([1,4,9,16]), linewidth = 2.0) # it takes lists, numpy arrays all
# first is x axis , second list is y -axis
plt.ylabel("Squares")
plt.xlabel("Numbers")
plt.title("Line Plot")
plt.grid() # shows in grid easy to visualize
plt.show()
```

```python
# use setp()

x1 = [1,2,3,4]
y1 = [1,4,9,16]
x2 = [1,2,4, 6, 8]
y2 = [2,4, 8, 12, 16]

lines = plt.plot(x1, y1, '--', x2, y2, '-') # These properties remain
plt.setp(lines[0], color = 'r', linewidth = 2.0) # Python style
plt.setp(lines[1], 'color' , 'g', 'linewidth' , 2.0) # Matlab style
plt.grid()
plt.show()
```

```python
# Play with multiple plots
np.random.seed(62)
x = np.linspace(0, 2*np.pi, 500)

plt.figure(1) # creates a new figure
plt.subplot(1,2,1) # 121 indicates 1 row, 2 columns, plot1 (left)
plt.plot(x, np.sin(x))
plt.grid()

plt.subplot(1,2,2)  # 122 indicates 1 row, 2 columns, plot2 (right)
plt.plot(x, np.cos(x))
plt.grid()
plt.show()

a = np.random.randint(1,100,20)
a = np.sort(a)
print(a)

plt.figure(2)
plt.plot(a)
plt.grid()
plt.show()
```
